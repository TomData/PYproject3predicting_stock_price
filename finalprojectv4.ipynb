```python
# 8. Create/Get financial_ratios: P/E (price close/sum of last 4 quarters of eps (not diluted))

# 8.1 create stock_income_balance_cash df with shorter name to work with
sibc = stock_income_balance_cash

# 8.2 only select relevant columns
sibc_rel = sibc.iloc[:,[0,1,2,3,4,34]]

# 8.3 create df with non null eps
eps_df = sibc_rel[sibc_rel['eps'].notnull()]
print(eps_df.head(5))

# 8.4 create rolling sum with window = 4
eps_df.loc[:,'epstrailingfourquartersSum'] = eps_df.loc[:,'eps'].rolling(window=4).sum()
print(eps_df.head(5))

# 8.5 create new column 'epstrailingfourquartersSum' for daily dataframe sibc
sibc.loc[:,'epstrailingfourquartersSum'] = np.zeros(len(sibc))
print(sibc[['Date','eps']].head(5))

# 8.6 loop through the eps df
for i in range(len(eps_df[eps_df['epstrailingfourquartersSum'].notnull()])-1):
    print(range(len(eps_df[eps_df['epstrailingfourquartersSum'].notnull()])-1))
    #get start and end indexes from eps dataframe
    start_index = eps_df.index[3+i]
    print(start_index)
    end_index = eps_df.index[3+i+1] -1#to be excluded 
    print(end_index)
    #modify daily dataframe d
    sibc.loc[start_index:end_index,'epstrailingfourquartersSum'] = [eps_df["epstrailingfourquartersSum"][eps_df.index[3+i]]]*(end_index-start_index+1)
    print([eps_df["epstrailingfourquartersSum"][eps_df.index[3+i]]]*(end_index-start_index+1))

# 8.7 fill the last part 
sibc.loc[end_index-1:,'epstrailingfourquartersSum'] = [eps_df["epstrailingfourquartersSum"][eps_df.index[3+i]]]*(len(sibc)-end_index+1)
print(sibc[['Date','eps','epstrailingfourquartersSum']].tail(5))

# 8.8 create new column 'p/e'
sibc.loc[:,'p/e(close)'] = sibc.loc[:,'Close']/sibc.loc[:,'epstrailingfourquartersSum']
print(sibc[['Date','eps','epstrailingfourquartersSum','p/e(close)']].tail(5))

# 8.9 save file

#sibc.to_csv(r'C:\Users\Thomas S\Documents\Ubiqum\5. Data Science\Final Project\stockdatav2.csv', index = True)
```

```python
# 9. Create/Get financial_ratios: P/B (price close (daily)/(total assets(quarterly)- goodwill and intangible assets(quaretly))/weighted average shares outstainding diluted

# 9.1 create new column: totalAssets_daily
sibc["totalAssets_daily"] = np.zeros(len(sibc))

#initialise value
value = np.nan

#loop through to create daily Asset column
for r in range(len(sibc)):    
    if pd.notnull(sibc.loc[r, "totalAssets"]):
        value = sibc.loc[r, "totalAssets"]
        #print(value,sibc.loc[r,"totalAssets_daily"])
        
    sibc.loc[r,"totalAssets_daily"] = value
    #print(value,df.loc[r,"totalAssets_daily"])

        
# 9.2 create new column: goodwillAndIntangibleAssets_daily
sibc["goodwillAndIntangibleAssets_daily"] = np.zeros(len(sibc))
        
#initialise value
value1 = np.nan

#loop through to create daily goodwill and intangible asset column
for r in range(len(sibc)):    
    if pd.notnull(sibc.loc[r, "goodwillAndIntangibleAssets"]):
        value1 = sibc.loc[r, "goodwillAndIntangibleAssets"]
        #print(value1,sibc.loc[r,"goodwillAndIntangibleAssets_daily"])  
        
    sibc.loc[r,"goodwillAndIntangibleAssets_daily"] = value1
    #print(value1,df.loc[r,"goodwillAndIntangibleAssets_daily"])
        
# 9.3 create new column: weightedAverageShsOutDil_daily
sibc["weightedAverageShsOutDil_daily"] = np.zeros(len(sibc))
        
#initialise value
value2 = np.nan

#loop through to create daily goodwill and intangible asset column
for r in range(len(sibc)):    
    if pd.notnull(sibc.loc[r, "weightedAverageShsOutDil"]):
        value2 = sibc.loc[r, "weightedAverageShsOutDil"]
        #print(value2,sibc.loc[r,"goodwillAndIntangibleAssets_daily"])  
        
    sibc.loc[r,"weightedAverageShsOutDil_daily"] = value2
    #print(value2,df.loc[r,"goodwillAndIntangibleAssets_daily"])

# 9.4 create pb ratio of the newly created daily values
sibc.loc[:,'p/b'] = sibc.loc[:,'Close']/((sibc.loc[:,'totalAssets_daily']-sibc.loc[:,'goodwillAndIntangibleAssets_daily'])/sibc.loc[:,'weightedAverageShsOutDil_daily'])
print(sibc[['Close','totalAssets_daily','goodwillAndIntangibleAssets_daily','weightedAverageShsOutDil_daily', 'p/b']].tail(5))

# 9.5 save file
sibc.to_csv(r'C:\Users\Thomas S\Documents\Ubiqum\5. Data Science\Final Project\stockdatav3.csv', index = True)   
```

```python
# Compute FUTURE VALUES & Growth rates CHANGE NWC & FCF

#compute percent change
all_yearly.insert(31,'pct_ChangeWC_change',all_yearly[['ChangeWC4Qsum']].pct_change())

# weighted yearly growth/decline of the "computed percent change" (is used to predict the 5Y percent change which is used to compute the 5Y predicted values)
all_yearly.insert(32,'pct_ChangeWC_change_change',all_yearly[['pct_ChangeWC_change']].multiply(100).pct_change())
#print(all_yearly)

#compute weighted avg. of the change of change (used to predict the future change & therefore future values)
weights = [0.25, 0.75]
data_rev = all_yearly.tail(5).iloc[3:5,32].values*.1
#print(data_rev)

# weighted avg. values used to calcualte future growth rates (which are used to calcualte future values)
ChangeWC_growth_multiplier = 1 + np.average(data_rev,weights = weights) #instead of constant value it would be better to use multiplier as part of string beginning with multiplier which going towards 0
print(ChangeWC_growth_multiplier)

#insert new column of forecast/predicted growth rate
all_future_yearly.insert(20,'changeNWC_growth', np.nan)

#add growth rate of most recent year
all_future_yearly.iloc[0,20] = all_yearly.tail(1).iloc[:,31].values
#print(all_future_yearly)

#fill the column of the forecast/predicted growth rates
for i in range(1,len(all_future_yearly)):
    all_future_yearly.loc[i, 'changeNWC_growth'] = all_future_yearly.loc[i-1, 'changeNWC_growth'] * ChangeWC_growth_multiplier

#fill the column of the forecast/predicted values
for i in range(1,len(all_future_yearly)):
    all_future_yearly.loc[i, 'changeNWC'] = all_future_yearly.loc[i-1, 'changeNWC'] * (1+all_future_yearly.loc[i,'changeNWC_growth'])

#calcualte FCF
all_future_yearly.loc[:,'freecashflow'] = all_future_yearly['netincome'] + all_future_yearly['DandA'] - all_future_yearly['CAPX'] - all_future_yearly['changeNWC']

print(all_future_yearly)
```

```python
# Compute Discount Factors
all_future_yearly['discount_factors'] = np.nan
#print(all_future_yearly)

for i in range (0,len(all_future_yearly)):
    all_future_yearly.loc[i,'discount_factors'] = 1/(1+WACC)**i
    
#print(all_future_yearly)

# Compute Discounted future free cash flows
all_future_yearly['discounted_ffcf'] = np.nan

for i in range (0, len(all_future_yearly)):
    all_future_yearly.loc[i,'discounted_ffcf'] = all_future_yearly.loc[i,'freecashflow'] * all_future_yearly.loc[i,'discount_factors']
    
print(all_future_yearly)

# Compute Discounted Terminal Value
terminal_growth_rate = 0.02
WACC = 0.0547 #https://www.gurufocus.com/term/wacc/NAS:MSFT/WACC-/Microsoft

terminal_value = (all_future_yearly.tail(1).iloc[:,21].values*(1+terminal_growth_rate))/(WACC-terminal_growth_rate)
print(terminal_value)

discount_factor_tv = 1/(1+WACC)**len(all_future_yearly)
print(discount_factor_tv)

dis_terminal_value = terminal_value * discount_factor_tv
print(dis_terminal_value)

# Compute Discounted Firm Value
dis_firm_value = all_future_yearly.loc[:,'discounted_ffcf'].values.sum() + dis_terminal_value
print(dis_firm_value)

# Get Current Debt Value (total liabilities)
total_libailities = all_yearly.tail(1).loc[:,'totalLiabilities'].values
print(total_libailities)

# Compute total quity value 
equity = dis_firm_value - total_libailities
print(equity)

# Get no. shares
shares = sibc[['weightedAverageShsOutDil']].loc[sibc.weightedAverageShsOutDil.notnull()].tail(1).values
print(shares)

#Compute share price
price = equity/shares
print(price)
```
